```python
#  用的极多的子图的画法
# 画子图的步骤
# 1. 把画板放大
# figsize 画板的宽度, 画板的高度, 单位是英寸
# 子图的列数决定宽度, 子图的行数决定高度
x = np.linspace(-10, 10, 100)
plt.figure(figsize=(3 * 4, 4 * 1))
# 2, 创建子图的轴面, 几行几列第几个
axes1 = plt.subplot(1, 3, 1)
# 3, 在子图中绘制你需要的图形
axes1.plot(x, np.sin(x))
axes1.grid(lw=2)

axes2 = plt.subplot(1, 3, 2)
axes2.plot(x, np.cos(x))
axes2.grid(lw=2)

axes3 = plt.subplot(1,3, 3)
axes3.plot(x ,x**2)
axes3.grid(c='g', ls='--')
```

### 坐标轴界限

#### axis方法
如果axis方法没有任何参数，则返回当前坐标轴的上下限axis(xmin =,ymax = )

```python
x = np.linspace(0, 2*np.pi, 100)
plt.plot(x, np.sin(x))
# axis 不传参数,返回当前的坐标轴界限
# 可以给axis传参,设置当前坐标轴的界限
# x轴的最小值,x轴的最大值,y轴的最小值,y轴的最大值
# plt.axis((0, 2 * np.pi, -1, 1))
# 如果指向设置某个界限, 可以传关键字参数
plt.axis(ymin=-1, ymax=1)
# matplotlib内置了一些axis轴线风格, 默认风格叫做tight,
# equal指x,y坐标轴宽度相同
plt.axis('equal')
```

### 保存图片

figure.savefig的选项
+ filename  
含有文件路径的字符串或Python的文件型对象。图像格式由文件扩展名推断得出，例如，.pdf推断出PDF，.png推断出PNG
（“png”、“pdf”、“svg”、“ps”、“eps”……）
+ dpi  
图像分辨率（每英寸点数），默认为100
+ facecolor  
图像的背景色，默认为“w”（白色） 

```python
# 通过figure对象把图片保存到本地
# 
figure = plt.figure()
plt.plot(x, y1, label='fast')
plt.plot(x, y2, label='normal')
plt.plot(x, y3, label='slow')
plt.legend(loc=(0, 1), ncol=3)
# 图片的后缀必须是当前市面上支持的图片后缀,
# dpi每英寸的点数, 就是图片的分辨率,dpi越大,图片越大, 图片的分辨率越大
figure.savefig('line.png', dpi=200, facecolor='g', transparent='True')
```

### X、Y轴坐标刻度

```python
plt.plot(x, y)
# 设置刻度, 第一个参数是刻度,第二个参数是刻度的标签,如果不设置标签,会自动用刻度作为标签.
_ = plt.xticks([0, np.pi/2, np.pi, 3*np.pi/2, 2*np.pi], ['0', '$\pi/2$', r'$\alpha$', '$3\pi/2$', '$2\pi$'])
# y轴刻度
_ = plt.yticks([-1, 0, 1])
```

#### 面向对象方法
set_xticks、set_yticks、set_xticklabels、set_yticklabels方法

```python
# 当在子图中设置刻度, 不太一样.
axes = plt.subplot()
axes.plot(x, y)
axes.set_xlabel('X')
axes.set_ylabel('y')
axes.set_title('Sin')

# 设置刻度不一样,子图里面set_xticks只能设置刻度,不能设置刻度的标签.
axes.set_xticks([0, np.pi/2, np.pi, 3*np.pi/2, 2*np.pi])
# 设置刻度的标签需要使用set_xticklabels
_ = axes.set_xticklabels(['0', '$\pi/2$', r'$\pi$', '$3\pi/2$', '$2\pi$'])
```
