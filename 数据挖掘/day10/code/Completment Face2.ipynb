```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
from sklearn.datasets import fetch_olivetti_faces
```

```python
faces = fetch_olivetti_faces()
```

```python
print(faces.DESCR)
```

```python
data = faces.data
images = faces.images
```

```python
data.shape
```

```python
# 取出训练数据和标记结果.
# 上半部分脸作为训练数据, 下半部分脸作为标记结果.
# plt.imshow(data[0, :2048].reshape(32, 64), cmap='gray')
face_up = data[:, :2048]
face_down = data[:, 2048:]
```

```python
plt.imshow(face_down[0].reshape(32, 64), cmap='gray')
```

生成训练数据和预测数据

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(face_up, face_down, test_size=0.2)
```

创建机器学习模型，以字典的形式包含四种模型

```python
# knn, ridge, linear, tree
from sklearn.neighbors import KNeighborsRegressor
from sklearn.linear_model import LinearRegression, Ridge
from sklearn.ensemble import RandomForestRegressor
```

```python
from sklearn.ensemble import ExtraTreesRegressor
```

```python
knn = KNeighborsRegressor()
knn.fit(X_train, y_train)
```

```python
# 使用字典,同时去训练
estimators = {
    'knn': KNeighborsRegressor(),
    'ridge': Ridge(),
    'linear': LinearRegression(),
#     'tree': ExtraTreesRegressor()
}
```

```python
X_train.shape
```

```python
y_preds = {}
for name, estimator in estimators.items():
    estimator.fit(X_train, y_train)
    y_ = estimator.predict(X_test)
    y_preds[name] = y_
    # 打印得分
    print(name, estimator.score(X_test, y_test))
```

```python
y_knn = y_preds['knn']
```

```python
plt.imshow(X_test[0].reshape(32, 64), cmap='gray')
```

```python
face_stack = np.vstack((X_test[0].reshape(32, 64), y_knn[0].reshape(32, 64)))
plt.imshow(face_stack, cmap='gray')
```

```python
# 8行4列, 第一列是真实的人脸数据, 第二列到最后列是各个算法预测的下半部分脸和真实的上半部脸拼起来的人脸图片
plt.figure(figsize=(4 * 3, 8 * 3))
for i in range(8):
   # 第一列数据
    axes = plt.subplot(8, 4, i * 4 + 1)
    face_up = X_test[i * 10].reshape(32, 64)
    face_down = y_test[i * 10].reshape(32, 64)
    true_face = np.vstack((face_up, face_down))
    axes.imshow(true_face, cmap='gray')
    axes.axis('off')
    if i == 0:
        axes.set_title('True Face')
        
    for j, name in enumerate(y_preds.keys(), 1):
        y_pred = y_preds[name]
        axes = plt.subplot(8, 4, i * 4 + 1 + j)
        pred_face_down = y_pred[i * 10].reshape(32, 64)
        pred_face = np.vstack((face_up, pred_face_down))
        axes.imshow(pred_face, cmap='gray')
        axes.axis('off')
        if i == 0:
            axes.set_title(name)
```
