```python
from sklearn.neighbors import KNeighborsRegressor
```

```python
plt.figure(figsize=(3*10,13*5))
for i, col in enumerate(df.columns[0:-1],1):
    X_train = df[[col]]
    X_test = np.linspace(X_train.min(), X_train.max(),100).reshape(-1,1)
    
    knn = KNeighborsRegressor()
    knn.fit(X_train, df.iloc[:, -1])
    yk_ = knn.predict(X_test)
    
    linear = LinearRegression()
    linear.fit(X_train, df.iloc[:, -1])
    y_ = linear.predict(X_test)
    
    ridge = Ridge()
    ridge.fit(X_train, df.iloc[:, -1])
    y_r = ridge.predict(X_test)
    
    lasso = Lasso()
    lasso.fit(X_train, df.iloc[:, -1])
    y_l = lasso.predict(X_test)
    
    poly2 = PolynomialFeatures()
    X_train_poly2 = poly2.fit_transform(X_train)
    linear_poly2 = LinearRegression()    
    linear_poly2.fit(X_train_poly2, df.iloc[:, -1])
    X_test_poly2 = poly2.fit_transform(X_test)
    y2 = linear_poly2.predict(X_test_poly2)
     
    
    
    axes = plt.subplot(5,3,i)
    axes.scatter(X_train, df.iloc[:, -1], c='#CEF6E3')
    axes.plot(X_test,yk_,c='b')
    axes.plot(X_test,y_,c='r')
    axes.plot(X_test,y_r,c='g')
    axes.plot(X_test,y_l,c='y')
    axes.plot(X_test,y2,c='gray')
    score = linear.score(X_train, df.iloc[:, -1])
    score2 = ridge.score(X_train, df.iloc[:, -1])
    score3 = ridge.score(X_train, df.iloc[:, -1])
    score4 = linear_poly2.score(X_train_poly2, df.iloc[:, -1])
    axes.set_title(f'{col}\n score:{score}\n score2:{score2}\n score3:{score3}\n score4:{score4}',fontdict=dict(size=20,color='r'))
```

# 人脸作业

```python
from sklearn.datasets import fetch_olivetti_faces
```

```python
faces = fetch_olivetti_faces()
```

```python
faces
```

```python
data = faces.data
images = faces.images
```

```python
plt.imshow(images[6], cmap='gray')
```

```python
images[0].shape
```

```python
data.shape
```

```python
face_left = images[:, :, :32]
face_right = images[:, :, 32:]
```

```python
plt.imshow(face_right[0], cmap='gray')
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(face_left, face_right, test_size=0.2)
```

```python
# knn, ridge, linear, tree
from sklearn.neighbors import KNeighborsRegressor
from sklearn.linear_model import LinearRegression, Ridge
from sklearn.ensemble import RandomForestRegressor
```

```python
from sklearn.ensemble import ExtraTreesRegressor
```

```python
estimators = {
    'knn': KNeighborsRegressor(),
    'ridge': Ridge(),
    'linear': LinearRegression(),
#     'tree': ExtraTreesRegressor()
}
```

```python
y_preds = {}
for name, estimator in estimators.items():
    estimator.fit(X_train.reshape(-1, 2048), y_train.reshape(-1, 2048))
    y_ = estimator.predict(X_test.reshape(-1, 2048))
    y_preds[name] = y_
    # 打印得分
    print(name, estimator.score(X_test.reshape(-1, 2048), y_test.reshape(-1, 2048)))
```

```python
y_knn = y_preds['knn']
```

```python
face_stack = np.hstack((X_test[0].reshape(64,32), y_knn[0].reshape(64,32)))
plt.imshow(face_stack, cmap='gray')
```

```python
# 8行4列, 第一列是真实的人脸数据, 第二列到最后列是各个算法预测的下半部分脸和真实的上半部脸拼起来的人脸图片
plt.figure(figsize=(4 * 3, 8 * 3))
for i in range(8):
   # 第一列数据
    axes = plt.subplot(8, 4, i * 4 + 1)
    face_left = X_test[i * 10]
    face_right = y_test[i * 10]
    
    true_face = np.hstack((face_left, face_right))
    axes.imshow(true_face, cmap='gray')
    axes.axis('off')
    if i == 0:
        axes.set_title('True Face')
        
    for j, name in enumerate(y_preds.keys(), 1):
        y_pred = y_preds[name]
        axes = plt.subplot(8, 4, i * 4 + 1 + j)
        pred_face_right = y_pred[i * 10]
        pred_face = np.hstack((face_left.reshape(64,32), pred_face_right.reshape(64,32)))
        axes.imshow(pred_face, cmap='gray')
        axes.axis('off')
        if i == 0:
            axes.set_title(name)
```
