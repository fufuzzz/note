```python
# 对售价进行区分
bins = [0, 50, 100, 150, 200, 500]
labels = ['0-50', '50-100','100-150', '150-200','200以上']
derat['价格区间'] = pd.cut(derat['售价'], bins=bins, labels=labels)
```

```python
def convert(x):
    return 1 - (x - x.min()) / (x.max() - x.min())
```

```python
def process_data(df, by,order_by='相对竞争度', ascending=True):
    collection = df.groupby(by=by)[['预估销售额']].sum()
    collection['销售额占比'] = collection['预估销售额'] / collection['预估销售额'].sum()
    collection['宝贝数'] = df.drop_duplicates(subset=['宝贝ID']).groupby(by=by)['宝贝ID'].count()
    collection['宝贝数占比'] = collection['宝贝数'] / collection['宝贝数'].sum()
    collection['单宝贝平均销售额'] = collection['预估销售额'] / collection['宝贝数']
    collection['相对竞争度'] =  collection['单宝贝平均销售额'].transform(convert)
    # 根据相对竞争度排序
    collection.sort_values(by=order_by, inplace=True)
    return collection
```

```python
def draw_collection(collection):
    plt.figure(figsize=(12, 5))
    collection['销售额占比'].plot(kind='bar', color='wheat')
    collection['相对竞争度'].plot(marker='o', rot=90)
    plt.ylabel('销售额占比')
```

```python
process_data(derat, by='价格区间')
```

```python
draw_collection(process_data(derat, by='价格区间'))
```

```python
# 先分析拜耳的产品结构
ansu_prod = pd.read_excel(files[0])
ansu_prod.head()
```

```python
ansu_prod.info()
```

```python
ansu_prod.drop_duplicates(subset=['商品']).shape   # 49种商品
```

```python
def data_process(df):
    df_total = df.groupby(by=['商品'])[['交易增长幅度']].mean()
    df_total['交易金额'] = df.groupby(by=['商品'])['交易金额'].sum()
    df_total['交易金额占比'] = df_total['交易金额'] / df_total['交易金额'].sum()
    df_total['商品个数'] = df.groupby(by=['商品'])['操作'].count()
    return df_total
```

```python
ansu_prod_total = data_process(ansu_prod.copy())
```

```python
def mask(s, q=0.9):
    threshold = s.quantile(q)
    s.loc[s >= threshold] = threshold
    return s
```

```python
ansu_prod_total['交易增长幅度'] = mask(ansu_prod_total['交易增长幅度'].copy())
ansu_prod_total['交易金额占比'] = mask(ansu_prod_total['交易金额占比'].copy())
```

```python
ansu_prod_total.describe()
```

```python
# ansu_prod_total['交易金额占比'] = ansu_prod_total['交易金额'] / ansu_prod_total['交易金额'].sum()
# ansu_prod_total.describe()
```

```python
ansu_prod_total.reset_index(inplace=True)
```

```python
def sort_products(df_total, sort):
    if sort == '明星类':
        return df_total.loc[(df_total['交易金额占比'] > df_total['交易金额占比'].quantile(0.5)) & (df_total['交易增长幅度'] > df_total['交易增长幅度'].quantile(0.5))]
    elif sort == '奶牛类':
        return df_total.loc[(df_total['交易金额占比'] > df_total['交易金额占比'].quantile(0.5)) & (df_total['交易增长幅度'] < df_total['交易增长幅度'].quantile(0.5))]
    elif sort == '问题类':
        return df_total.loc[(df_total['交易金额占比'] < df_total['交易金额占比'].quantile(0.5)) & (df_total['交易增长幅度'] > df_total['交易增长幅度'].quantile(0.5))]
```

```python
sort_products(ansu_prod_total, '明星类')
```

```python
sort_products(ansu_prod_total, '奶牛类')
```

```python
sort_products(ansu_prod_total, '问题类')
```

```python
def drow_boston_matrix(df_total):   
    plt.figure(figsize=(8, 6))
    sns.scatterplot(x='交易金额占比', y='交易增长幅度', data=df_total, palette='cool', hue='商品个数')
    plt.axhline(y=df_total['交易增长幅度'].median())
    plt.axvline(x=df_total['交易金额占比'].median())
#     plt.xlim(-0.01, 0.04)
    for i in df_total.index:
        plt.text(df_total.iloc[i, 3] + 0.001, df_total.iloc[i, 1] + 0.1, i)
```

```python
drow_boston_matrix(ansu_prod_total)
plt.xlim(-0.01, 0.045)
```
