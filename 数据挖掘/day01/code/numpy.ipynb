```python
n1 = np.random.randint(0, 10, size=(3, 5))
n2 = np.random.randint(0, 10, size=(3, 5))
```

```python
# 垂直级联
np.concatenate((n1, n2), axis=0)
```

### 2. 广播机制

【重要】ndarray广播机制的两条规则
- 规则一：为缺失的维度补1
- 规则二：假定缺失元素用已有值填充


例1：
m = np.ones((2, 3))
a = np.arange(3)
求M+a

```python
m = np.ones((2, 3))
a = np.arange(3)
display(a, m)
```

```python
# numpy有个广播机制, 广播机制的目的就是为了把两个形状不容的nadarray变成形状相同的ndarray一遍他们可以进行运算
m + a
```

```python
a 缺维度, 给a补维度.
a [0, 1, 2] -> a [[0, 1, 2]]
a缺数据,给a用已有的数据进行填充
a [[0, 1, 2]] -> a[[0, 1, 2], [0, 1, 2]]
```

```python
# 广播机制的规则:
1. 缺维度,补维度
2. 缺数据,补数据
```
