假设有两份成绩单，除了ddd是张三李四王老五之外，还有ddd4是张三和赵小六的成绩单，如何合并？

如果ddd4中张三的名字被打错了，成为了张十三，怎么办？

自行练习多对一，多对多的情况

自学left_index,right_index

```python
# 假设有两份成绩单，除了ddd是张三李四王老五之外，还有ddd4是张三和赵小六的成绩单，
# 如何合并？
columns = ['name', '语文', '数学']
data = np.random.randint(0, 150, size=(3, 3))
ddd5 = DataFrame(data=data, index=list('012'), columns=columns)
ddd5['name'] = ['张三', '李四', '王五']
data2 = np.random.randint(0, 150, size=2)
ddd6 = DataFrame({'name':['张三','赵小六'],'语文':data2,'数学':data2})
ddd5.iloc[0] = ddd6.iloc[0]
display(ddd5, ddd6)
```

```python
ddd5.merge(ddd6, how='outer')
```

```python
ddd6.iloc[0, 0] = '张十三'
display(ddd5, ddd6)
```

```python
ddd5.merge(ddd6,on=('语文','数学'), how='outer')
```

### 考虑应用情景，使用多种方式合并ddd与ddd4

```python
df1 = DataFrame({'name':['张三','李四','Chales'],'id':[1,2,3],'age':[22,21,25]})

df2 = DataFrame({'sex':['男','男','女'],'id':[2,3,4],'group':['sale','search','service']})
display(df1, df2)
```

```python
# 一对一合并
df1.merge(df2)
```

```python
df1 = DataFrame({'name':['张三','李四','Chales'],'id':[1,2,2],'age':[22,21,25]})

df2 = DataFrame({'sex':['男','男','女'],'id':[2,3,4],'group':['sale','search','service']})
display(df1,df2)
```

```python
# 多对一合并
df1.merge(df2)
```

```python
df1 = DataFrame({'name':['张三','李四','张三'],'salary':[10000,12000,20000],'age':[22,21,25]})

df2 = DataFrame({'sex':['男','男','女'],'name':['张三','张三','凡凡'],'group':['sale','search','service']})
display(df1, df2)
```

```python
# 多对多合并
df1.merge(df2)
```

### 假设有两个同学都叫李四，ddd5、ddd6都是张三和李四的成绩表，如何合并？

```python
ddd5 = DataFrame({'name':['张三','李四','李四'],'degree':[110,150,89]})
ddd6 = DataFrame({'name':['张三','李四','李四'],'degree':[119,149,77]})
display(ddd5,ddd6)
```

```python
ddd5.merge(ddd6,left_index=True,right_index=True,suffixes=('期中', '期末'))
```

```python
ddd5.merge(ddd6,on='name', suffixes=('期中', '期末'))
```

```python
ddd5.merge(ddd6,on='name')
```
