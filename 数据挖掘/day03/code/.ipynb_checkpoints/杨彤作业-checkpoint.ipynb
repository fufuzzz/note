练习16：

假设有两份成绩单，除了ddd是张三李四王老五之外，还有ddd4是张三和赵小六的成绩单，如何合并？

如果ddd4中张三的名字被打错了，成为了张十三，怎么办？

自行练习多对一，多对多的情况

自学left_index,right_index

============================================

```python
index = [0, 1,2]
columns = ['name','语文', '数学', '英语']
data = np.random.randint(0,150,size=(3,4))
ddd = DataFrame(data=data,index=index,columns=columns)
ddd['name']=['张三','李四','王老五']
ddd
```

```python
index = [0, 1]
columns = ['name','语文', '数学', '英语']
data = np.random.randint(0,150,size=(2,4))
ddd4 = DataFrame(data=data,index=index,columns=columns)
ddd4['name']=['张三','赵小六']
ddd4.iloc[0]=ddd.iloc[0]
ddd4
```

```python
ddd.merge(ddd4,how='outer')
```

```python
# ddd.merge(ddd4,left_index=True,right_index=True)
```

```python
index = [0, 1]
columns = ['name','语文', '数学', '英语']
data = np.random.randint(0,150,size=(2,4))
ddd4 = DataFrame(data=data,index=index,columns=columns)
ddd4['name']=['张十三','赵小六']
ddd4.iloc[0]=ddd.iloc[0]
ddd4
```

```python
ddd.merge(ddd4)
```

1) 一对一合并

```python
df1 = DataFrame({'name':['张三','李四','Chales'],'id':[1,2,3],'age':[22,21,25]})

df2 = DataFrame({'sex':['男','男','女'],'id':[2,3,4],'group':['sale','search','service']})
display(df1, df2)
```

```python
df1.merge(df2)
```

2) 多对一合并

```python
df1 = DataFrame({'name':['张三','李四','Chales'],'id':[1,2,2],'age':[22,21,25]})

df2 = DataFrame({'sex':['男','男','女'],'id':[2,3,4],'group':['sale','search','service']})
display(df1,df2)
```

```python
df1.merge(df2) 
```

3) 多对多合并

```python
df1 = DataFrame({'name':['张三','李四','张三'],'salary':[10000,12000,20000],'age':[22,21,25]})

df2 = DataFrame({'sex':['男','男','女'],'name':['张三','张三','凡凡'],'group':['sale','search','service']})
display(df1, df2)
```

```python
df1.merge(df2)
```

练习17：

### 考虑应用情景，使用多种方式合并ddd与ddd4
============================================

```python
df1 = DataFrame({'name':['张三','李四','张三'],'salary':[10000,12000,20000],'age':[22,21,25]})

df2 = DataFrame({'age':[21,18,29],'名字':['张三','张三','凡凡'],'group':['sale','search','service']})
display(df1, df2)
```

```python
df1.merge(df2)
```

```python
df1.merge(df2,how='outer')
```

```python
df1.merge(df2,left_on='name',right_on='名字')
```

```python
df1.merge(df2, left_index=True, right_index=True)
```

```python
df1.merge(df2,left_on='name',right_on='名字',how='outer')
```

```python
df1.merge(df2,left_on='name',right_on='名字',how='left')
```

```python
df1.merge(df2,left_on='name',right_on='名字',how='right')
```

练习18：

### 假设有两个同学都叫李四，ddd5、ddd6都是张三和李四的成绩表，如何合并？
============================================

```python
ddd5 = DataFrame({'name':['张三','李四','李四'],'degree':[120,118,149]})
ddd6 = DataFrame({'name':['张三','李四','李四'],'degree':[100,101,102]})
display(ddd5,ddd6)
```

```python
ddd5.merge(ddd6,on='name',suffixes=('_期中', '_期末'))
```

```python
ddd5.merge(ddd6,left_index=True,right_index=True,suffixes=('_期中', '_期末'))
```
