```python
%matplotlib inline
import numpy as np
import pandas as pd
import seaborn as sns
from scipy import stats
import matplotlib as mpl
import matplotlib.pyplot as plt

sns.set(style="ticks")
```

```python
sns.set(style="whitegrid", color_codes=True)
```

```python
heart = pd.read_csv('heart.csv')
```

```python
heart.head()
```

```python
heart.info()
```

```python
# 查看正负样本的数量
sns.countplot(x=heart['target'], data=heart)
```

```python
plt.pie(heart['target'].value_counts(), autopct='%.2f%%')
```

```python
# 绘制热力图
corr = heart.corr()
corr
```

```python
plt.figure(figsize=(12, 12))
cmap = sns.diverging_palette(230, 20, as_cmap=True)
sns.heatmap(corr, annot=True, fmt='.2f', square=True, linewidths=0.5, cmap=cmap)
```

```python
# age是连续性特征.
# 对于连续型特征, 先看分布情况
sns.distplot(x=heart['age'])
```

```python
# 年龄在40有个小高峰, 60左右是最多的
sns.boxplot(y='age',data=heart)
```

```python
plt.rcParams['font.sans-serif'] = 'SimHei'
```

```python
bins = [20, 25, 30, 35, 40, 45, 50, 55, 60, 65, 70, 75, 80]
labels = ['20-25', '25-30', '30-35', '35-40', '40-45', '45-50 ','50-55', '55-60', '60-65','65-70', '70-75', '75-80']
heart['年龄区间'] = pd.cut(heart['age'], bins=bins, labels=labels)
```

```python
sns.countplot(x='年龄区间', data=heart, hue='target')
```

```python
# 40-45 是一个危险期, 50-55也算是
```

```python
# 在男女性别中, 不同年龄的男女患病情况
sns.catplot(x='年龄区间', hue='target', col=
           'sex', data=heart, kind='count')
```

```python
sns.countplot(x='cp', data=heart, hue='target') 
```

```python
sns.catplot(x='cp', hue='target', col=
           'sex', data=heart, kind='count')
```

```python
# 再引入slope
sns.catplot(x='cp', hue='target', col=
           'sex', row='slope', data=heart, kind='count')
```
