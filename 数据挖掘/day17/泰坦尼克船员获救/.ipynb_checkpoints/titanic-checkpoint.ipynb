```python
# 导包 三剑客
import numpy as np
import pandas as pd
from pandas import DataFrame, Series
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# 读取数据
titanic = pd.read_csv('titanic_train.csv')
titanic.head()
```

```python
# 数据预处理
titanic.describe()
# 查看是否有空数据
titanic.isnull().any()
# age ,cabin, embarked存在空数据
# 年龄对训练有用,用年龄的中位数去替换
titanic['Age'] = titanic['Age'].fillna(value=titanic['Age'].median())
```

```python
# 性别是字符串,做一个转化
titanic.loc[titanic['Sex'] == 'male', 'Sex'] = 0
titanic.loc[titanic['Sex'] == 'female', 'Sex'] = 1
```

```python
# 处理Embarked的缺失值
# 查看SCQ哪个出现最多
(titanic['Embarked'] == 'S').sum()
# S 出现的最多, 把nan数据替换为S
titanic['Embarked'] = titanic['Embarked'].fillna('S')
titanic.loc[titanic['Embarked'] == 'S', 'Embarked'] = 0
titanic.loc[titanic['Embarked'] == 'C', 'Embarked'] = 1
titanic.loc[titanic['Embarked'] == 'Q', 'Embarked'] = 2
```

```python
# cabin缺失太多,不好处理,不加入训练数据中, 数据处理完毕.
titanic.isnull().any()
```

```python
# 使用线性回归,对存活概率做预测
from sklearn.linear_model import LinearRegression
# 使用交叉验证,减少取数据不平衡造成的问题.
from sklearn.model_selection import KFold
kf = KFold()
# 进行预测的列
predictors = ['Pclass', 'Sex', 'Age', 'SibSp', 'Parch', 'Fare', 'Embarked']

linear = LinearRegression()
# 定义一个列表用来存放每次预测的结果
predictions = []

for train, test in kf.split(titanic[predictors], titanic['Survived']):
    # 取出训练数据
    X_train = titanic[predictors].iloc[train]
    y_train = titanic['Survived'].iloc[train]
    linear.fit(X_train, y_train)
    # 预测
    X_test = titanic[predictors].iloc[test]
    y_ = linear.predict(X_test)
    predictions.append(y_)
    
```

```python
# 把所有的预测结果水平级联,结果肯定是一维的.
predictions = np.hstack(predictions)
```

```python
predictions.shape
```

```python
predictions
```

```python
# 我们认为预测的结果的数字的大小表示这个人获救的大小.设定一个阈值,大于0.5认为获救了,小于等于0.5 认为没获救
predictions[predictions > .5] = 1
predictions[predictions <= .5] = 0
# 算准确率 预测值和真实值做对比,然后求和 ,再除于总数
accuracy = (predictions == titanic['Survived']).sum() / len(predictions)
accuracy
```

```python
# 使用cross_val_score 计算交叉验证的准确率,一步到位
# 换个算法,使用逻辑斯蒂回归
from sklearn.model_selection import cross_val_score
from sklearn.linear_model import LogisticRegression
logistic = LogisticRegression()
# 需要传模型,训练数据, 对应的标记,
scores = cross_val_score(logistic, titanic[predictors], titanic['Survived'], cv=3)
# 3折,所以算出来三个得分
scores
```

```python
# 算平均得分, 比上面直接用线性回归高一丢丢.
scores.mean()
```

```python
# 接下来就是看看随机森林的准确率了,看看有没有提高.
# from sklearn.model_selection import cross_val_score
from sklearn.ensemble import RandomForestClassifier
rfc = RandomForestClassifier()
# 需要传模型,训练数据, 对应的标记,
scores = cross_val_score(rfc, titanic[predictors], titanic['Survived'], cv=3)
# 3折,所以算出来三个得分, 又提高了一点点.
scores.mean()
```

```python
# 加多点树, 100棵, 并且做一些剪枝
rfc = RandomForestClassifier(n_estimators=100, min_samples_split=4, min_samples_leaf=2)
# 需要传模型,训练数据, 对应的标记,
scores = cross_val_score(rfc, titanic[predictors], titanic['Survived'], cv=3)
# 3折,所以算出来三个得分, 又提高了一点点.
# 做了个剪枝之后,提高不少大约3% min_samples_split=4, min_samples_leaf=2
scores.mean()
```

```python
# 得分还不够.我们考虑增加特征.自己创造一些特征出来.
# 这里是难点.
# 增加一列加做familysize 把兄弟姐妹和父母小孩的数量全加起来 .
titanic['FamilySize'] = titanic['SibSp'] + titanic['Parch']
```

```python
titanic.head()
```

```python
# 再增加一列, 我们认为人的名字的长度和获救与否也是有关系的,所以增加一列名字长度.
titanic['NameLength'] = titanic['Name'].map(lambda x: len(x))
```

```python
titanic.head()
```

```python
# 我们认为一个人的头衔也可能会影响获救的情况.
# 我们把人名中间的title取出来.
# 定一个函数把名字中间的title取出来.
# 每个名字的title都是前面一个空格,后面是字母,再后面接一个点.用正则匹配出来.
import re
def get_title(name):
    title_search = re.search(' ([A-Za-z]+)\.', name)
    if title_search:
        return title_search.group(1)
titles = titanic['Name'].map(get_title)
```

```python
# 看一下都有哪些title
titles.value_counts()
# col表示上校, rev表示牧师, Mlle表示法国小姐, Major陆军少校, Sir爵士, Capt上尉, Countess女伯爵, Jonkheer无名贵族, Don阁下,尊称,Mme夫人 Dona葡萄牙语的夫人小姐
```

```python
titanic_test = pd.read_csv("test.csv")
titanic_test["Age"] = titanic_test["Age"].fillna(titanic["Age"].median())
titanic_test["Fare"] = titanic_test["Fare"].fillna(titanic_test["Fare"].median())
titanic_test.loc[titanic_test["Sex"] == "male", "Sex"] = 0 
titanic_test.loc[titanic_test["Sex"] == "female", "Sex"] = 1
titanic_test["Embarked"] = titanic_test["Embarked"].fillna("S")

titanic_test.loc[titanic_test["Embarked"] == "S", "Embarked"] = 0
titanic_test.loc[titanic_test["Embarked"] == "C", "Embarked"] = 1
titanic_test.loc[titanic_test["Embarked"] == "Q", "Embarked"] = 2
```

```python
titles = titanic_test['Name'].map(get_title)
titles.value_counts()
```

```python
# col表示上校, rev表示牧师, Mlle表示法国小姐, Major陆军少校, Sir爵士, Capt上尉, Countess女伯爵, Jonkheer无名贵族, Don阁下,尊称,Mme夫人 Dona葡萄牙语的夫人小姐
# 尽量把相同的意思的title归成一类.
title_mapping = {"Mr": 1, "Miss": 2, "Mrs": 3, "Master": 4, "Dr": 5, "Rev": 6, "Major": 7, "Col": 7, "Mlle": 8, "Mme": 8, "Don": 9, "Lady": 10, "Countess": 10, "Jonkheer": 10, "Sir": 9, "Capt": 7, "Ms": 2}
```

```python
# 增加一列title
for k,v in title_mapping.items():
    titles[titles == k] = v
    
titanic['Title'] = titles
```

```python
titanic.drop(columns='Title', inplace=True)
```

```python
# 用新的特征来做一下预测
predictors = ["Pclass", "Sex", "Age", "SibSp", "Parch", "Fare", "Embarked", "FamilySize", "Title", "NameLength"]
rfc = RandomForestClassifier(n_estimators=100, min_samples_split=4, min_samples_leaf=2)
# 需要传模型,训练数据, 对应的标记,
scores = cross_val_score(rfc, titanic[predictors], titanic['Survived'], cv=3)
# 3折,所以算出来三个得分, 又提高了一点点.
# 做了个剪枝之后,提高不少大约3% min_samples_split=4, min_samples_leaf=2
scores.mean()
# 只是提高了0.005% 提高不多啊. 再做一个特征选择吧.
```

```python
# 做特征选择使用SelectKBest
from sklearn.feature_selection import SelectKBest , f_classif
# 取5个主要特征
selector = SelectKBest(k=5)
selector.fit(titanic[predictors], titanic['Survived'])
```

```python
# 所有特征的得分,越高越好
selector.scores_
```

```python
# 所有特征的p值, 
selector.pvalues_
```

```python
selector.get_support(False) # 筛选出来的特征.
```

```python
# 把小数转化为分数
scores = -np.log10(selector.pvalues_)
# 然后画图
plt.bar(range(len(predictors)), scores)
plt.xticks(range(len(predictors)), predictors, rotation='vertical')
```
