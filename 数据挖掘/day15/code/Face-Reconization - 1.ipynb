```python
# 降维
pca = PCA(n_components=160, whiten=True)

# pca.fit(X_train)
# data_pca = pca.transform(X_train)

X_train_pca =  pca.fit_transform(X_train)
```

```python
from sklearn.svm import SVC
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
svc = SVC()
param_grid = {
    'C': [0.1, 1, 10, 20],
    'gamma': [1/160, 10/160, 100/160]
}
gv = GridSearchCV(svc, param_grid=param_grid, cv=5, verbose=1, n_jobs=-1)
```

```python
%time gv.fit(X_train_pca, y_train)
```

```python
def transform_img(img):
    img_gray = np.dot(img, [0.299, 0.587, 0.114])
    x = images[0].shape[0] / img_gray.shape[0]
    y = images[0].shape[1] / img_gray.shape[1]
    img_zoomed = ndimage.zoom(img_gray, zoom=(x,y))
    img_pca = pca.transform(img_zoomed.reshape(1, -1))
    return img_pca
```

```python
img = plt.imread('./bush2.png')
```

```python
gv.predict(transform_img(img))
```
