```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
from sklearn.datasets import make_circles
```

```python
data, target = make_circles(factor=0.2, noise=0.1)
plt.scatter(data[:, 0], data[:, 1], c=target)
plt.axis('equal')
```

```python
from sklearn.cluster import KMeans
```

```python
kmeans = KMeans(2)
kmeans.fit(data)
```

```python
data, target = make_circles(factor=0.2, noise=0.10)
plt.scatter(data[:, 0], data[:, 1], c=kmeans.labels_)
plt.axis('equal')
```

```python
kmeans.labels_
```

```python
# 数据有形状, kmeans聚类效果不好了.
# euclidean 欧氏距离
from sklearn.cluster import DBSCAN
```

```python
data
```

```python
dbscan = DBSCAN(eps=0.4, min_samples=4)
dbscan.fit(data, target)
plt.scatter(data[:,0], data[:, 1], c=dbscan.labels_)
plt.axis('equal')
```

```python
target
```

```python
# labels_中的-1表示离群点即噪声点.
dbscan.labels_
```

```python
from sklearn.metrics import silhouette_samples, silhouette_score
```

```python
silhouette_score(data, dbscan.labels_)
```

```python
silhouette_samples(data, dbscan.labels_)
```
