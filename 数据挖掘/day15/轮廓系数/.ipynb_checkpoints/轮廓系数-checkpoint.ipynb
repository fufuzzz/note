聚类算法的评估指标,轮廓系数

## 聚类评估：轮廓系数（Silhouette Coefficient ）

<img src="1.png" alt="FAO" width="490">

- 计算样本i到同簇其他样本的平均距离ai。ai 越小，说明样本i越应该被聚类到该簇。将ai 称为样本i的簇内不相似度。
- 计算样本i到其他某簇Cj 的所有样本的平均距离bij，称为样本i与簇Cj 的不相似度。定义为样本i的簇间不相似度：bi =min{bi1, bi2, ..., bik}


* si接近1，则说明样本i聚类合理
* si接近-1，则说明样本i更应该分类到另外的簇
* 若si 近似为0，则说明样本i在两个簇的边界上。

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
beer = pd.read_csv('./data.txt', sep='\s', engine='python')
beer
```

```python
# 计算轮廓系数
from sklearn.metrics import silhouette_samples, silhouette_score
```

```python
from sklearn.cluster import KMeans
```

```python
data = beer.iloc[:, 1:]
```

```python
kmeans = KMeans(3)
kmeans.fit(data)
```

```python
# 总体娥轮廓系数
silhouette_score(data, kmeans.labels_)
```

```python
# 每个样本的轮廓系数
silhouette_samples(data, kmeans.labels_).mean()
```

```python
ks = range(2, 20)
```

```python
# 通过计算;轮廓系数找到最佳的k值
scores = []
for k in ks:
    kmeans = KMeans(k)
    kmeans.fit(data)
    score = silhouette_score(data, kmeans.labels_)
    scores.append(score)
    print(k, score)
```

```python
for xy in zip(ks, scores):
    print(xy)
    
```

```python
plt.figure(figsize=(9, 8))
plt.plot(ks, scores, marker='o')
for xy in zip(ks, scores):
    plt.annotate("(%s,%s)" % xy, xy=xy)


# plt.xticks(ks)
# plt.yticks(scores)
# plt.grid()
```

```python
# 结论K取3比较合适
```

```python
# 对于数据范围差距比较大的数据, 对于有监督学习来说,我们会做标准化处理, 归一化处理
# 对于无监督, 能不能做类似的事情
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
data_scaled = scaler.fit_transform(data)
```

```python
data_scaled.std()
```

```python
data_scaled.mean()
```

```python
scores = []
for k in ks:
    kmeans = KMeans(k)
    kmeans.fit(data_scaled)
    score = silhouette_score(data_scaled, kmeans.labels_)
    scores.append(score)
    print(k, score)
```

```python
plt.figure(figsize=(9, 8))
plt.plot(ks, scores, marker='o')
for xy in zip(ks, scores):
    plt.annotate("(%s,%s)" % xy, xy=xy)
```

```python
# # 发现标准化处理并没有提高聚类的得分, 反而大幅降低了得分.
```

```python
# 记住: 对于聚类算法,不要做标准处理. 不利于kmeans根据距离进行聚类. 
```
